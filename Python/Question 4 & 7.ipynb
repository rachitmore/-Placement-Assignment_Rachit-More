Excepted Output Data Attributes

```python
import csv
import requests
import json
import numpy as np
import pandas as pd
```

```python
def download_data(url):
    response = requests.get(url)  # Send a GET request to the URL
    data = response.json()  # Convert the response to JSON format
    return data

# URL to download the data
url = r'https://data.nasa.gov/resource/y77d-th95.json'

# Download the data from the URL
data = download_data(url)
```

Sample Data

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.columns
```

```python
for feature in df.columns:
    print(feature,df[feature].unique())
```

```python
#Here we got to know that all there is only year feature has some random values need to fixed
```

```python
df['year'].iloc[37]
```

```python
df['year'].value_counts()
```

```python
#We have changed nan value to most frequent value
df['year'].iloc[37] = "1933-01-01T00:00:00.000"
```

```python
#converting 'year' feature to datetime format 
for i in range(df.shape[0]):
    df['year'].iloc[i] = pd.to_datetime(df['year'].iloc[i], errors = 'coerce')
```

```python
df.head()
```

```python
df['id'] = df['id'].astype("int")
df['mass'] = df['mass'].astype("float")
df['reclat'] = df['reclat'].astype("float")
df['reclong'] = df['reclong'].astype("float")
df['year'] = df['year'].astype('datetime64[ns]')
```

```python
df.head()
```

```python
print('name')
print("Required dtype-String")
meteorite_name = df['name']
print(meteorite_name.dtype)
print(meteorite_name[0])
print("")

print('id')
print("Required dtype-int")
meteorite_id = df['id']
print(meteorite_id.dtype)
print(meteorite_id[0])
print("")


print('nametype')
print("Required dtype-String")
meteorite_nametype=df['nametype']
print(meteorite_nametype.dtype)
print(meteorite_nametype[0])
print("")

print('recclass')
print("Required dtype-String")
meteorite_recclass=df['recclass']
print(meteorite_recclass.dtype)
print(meteorite_recclass[0])
print("")

print("mass")
print("Required dtype-Float")
meteorite_mass = df["mass"]
print(meteorite_mass.dtype)
print(meteorite_mass[0])
print("")


print("year")
print("Required dtype-Datetimeformat")
meteorite_year = df["year"]
print(meteorite_year.dtype)
print(meteorite_year[0])
print("")

print("reclat")
print("Required dtype-Float")
meteorite_reclat= df["reclat"]
print(meteorite_reclat.dtype)
print(meteorite_reclat[0])
print("")

print("reclong")
print("Required dtype -Float")
meteorite_reclong = df["reclong"]
print(meteorite_reclong.dtype)
print(meteorite_reclong[0])
print("")

print("geolocation")
print("Required dtype-list of int")
meteorite_geolocation = df["geolocation"]
print(meteorite_geolocation.dtype)
print(meteorite_geolocation[0])
print("")
```

```python
# Convert and save the data as a CSV file
csv_filename = "nasa_data.csv"
df.to_csv(csv_filename)
```
