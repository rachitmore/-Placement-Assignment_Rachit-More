```python
verbs_tag = ['WRB', 'VBZ', 'VBP', 'VBN', 'VBD', 'VBG', 'VB', 'RBR', 'RB', 'MD']
pronoun_tag = ['WP','PRP']
adjective_tag = ['JJS', 'JJR', 'JJ']
other_tag = ['WDT', 'DT', 'IN', 'CC', 'CD']
```

```python
def fetch_words(tags):
    #Creating blank series to concat series later
    Series = pd.Series("")
    for i in tags:
        df1 = df[df['pos_tag']==i]['word']
        Series = pd.concat([Series,df1])
    return Series
```

```python
verbs = fetch_words(verbs_tag)
pronouns = fetch_words(pronoun_tag)
adjectives = fetch_words(adjective_tag)
others = fetch_words(other_tag)
```
